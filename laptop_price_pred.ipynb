```python
X = df.drop(columns=['Price'])
y = np.log(df['Price'])
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.pipeline import Pipeline,make_pipeline
```

```python
df.head(1)
```

## Linear Regression

```python
step1 = ColumnTransformer(transformers=[
    ('col_tnf',OneHotEncoder(sparse_output=False,drop='first'),[0,1,2,3,5,8,10])
],remainder='passthrough')

step2 = LinearRegression()

pipe = Pipeline([
    ("col_trnf_ohe",step1),
    ("Model_linreg",step2)
])
```

```python
pipe.fit(X_train,y_train)
```

```python
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error as mse
from sklearn.metrics import mean_absolute_error as mae


y_pred = pipe.predict(X_test)

print('R2 score ', r2_score(y_pred,y_test))
print('MAE ', mae(y_pred,y_test))
```

## Ridge Regression

```python
step1 = ColumnTransformer(transformers=[
    ('col_tnf',OneHotEncoder(sparse_output=False,drop='first',dtype=int),[0,1,2,3,5,8,10])
],remainder='passthrough')

step2 = Ridge(alpha=10)

pipe_rr = Pipeline([
    ('step1',step1),
    ('step2',step2)
])

pipe_rr.fit(X_train,y_train)  ##


y_pred_rr = pipe_rr.predict(X_test)

print('R2 score ', r2_score(y_pred_rr,y_test))   ##
print('MAE ', mae(y_pred_rr,y_test))            ##
```

## Lasso Regression

```python
step1 = ColumnTransformer(transformers=[
    ('col_tnf',OneHotEncoder(sparse_output=False,drop='first',dtype=int),[0,1,2,3,5,8,10])
],remainder='passthrough')

step2 = Ridge(alpha=10)

pipe_lr = Pipeline([
    ('step1',step1),
    ('step2',step2)
])

pipe_lr.fit(X_train,y_train)


y_pred_lr = pipe_lr.predict(X_test)

print('R2 score ', r2_score(y_pred_lr,y_test))
print('MAE ', mae(y_pred_lr,y_test))
```
